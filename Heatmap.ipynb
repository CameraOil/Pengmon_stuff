```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Read from csv file
# take data from csv to 2D array

def csv_to_list(filename:str):
    data = []
    with open(file=filename, mode='r') as csv:
        lines = csv.readlines()
        rows = len(lines)
        print(rows)
        for i in range(0, -rows, -1):
            columns = [float(datum) for datum in lines[i-1].split(";")] # ";" bisa diganti "," tergantung file csv
            data.append(columns)
        
        csv.close()
        return data

    

    
        
```

```python
filename = "PSP kel 2.csv"
data_pengukuran = csv_to_list(filename)
print(data_pengukuran)

x_axis = [i for i in range(len(data_pengukuran[0]))]
y_axis = [i for i in range(len(data_pengukuran)-1,-1, -1)]
print(y_axis)

plt.figure(figsize=(10,10))
ax = sns.heatmap(data_pengukuran, annot=True, fmt = ".1f")
ax.set_xticklabels(x_axis)
ax.set_yticklabels(y_axis)
plt.title("TN3")
plt.xlabel("Horizontal")
plt.ylabel("Vertical")
plt.show()
```

```python
dataframe = pd.read_csv("PSP kel 2.csv")
dataframe.set_index('Titik', inplace=True)
dataframe
```

```python
# make the axis
# make the lists

x_ax = [i for i in range(1,9)]
y_ax = [i for i in range(1,13)]

titik = np.array(list(dataframe.index)).reshape((12, 8))
columns = list(dataframe.keys())

fig, axes = plt.subplots(3,2, figsize=(20,30))
i = 0
for y in range(2):
    for x in range(2):
        data = np.array(list(dataframe[columns[i]])).reshape((12,8))
        sns.heatmap(ax=axes[x,y], data=data, annot=titik)
        axes[x,y].set_title(columns[i])
        axes[x,y].set_xticklabels(x_ax)
        axes[x,y].set_yticklabels(y_ax)    
        i+=1

sns.heatmap(ax=axes[2,0],data = np.array(list(dataframe['Suara'])).reshape((12,8)),
            annot=titik)
axes[2,0].set_title('Suara')

    
plt.show()
```
