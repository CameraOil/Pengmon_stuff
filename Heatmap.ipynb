```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read from csv file
# take data from csv to 2D array

def csv_to_list(filename:str):
    data = []
    with open(file=filename, mode='r') as csv:
        lines = csv.readlines()
        rows = len(lines)
        print(rows)
        for i in range(0, -rows, -1):
            columns = [float(datum) for datum in lines[i-1].split(";")] # ";" bisa diganti "," tergantung file csv
            data.append(columns)
        
        csv.close()
        return data

    

    
        
```

```python
filename = "PSP kel 2.csv"
data_pengukuran = csv_to_list(filename)
print(data_pengukuran)

x_axis = [i for i in range(len(data_pengukuran[0]))]
y_axis = [i for i in range(len(data_pengukuran)-1,-1, -1)]
print(y_axis)

plt.figure(figsize=(10,10))
ax = sns.heatmap(data_pengukuran, annot=True, fmt = ".1f")
ax.set_xticklabels(x_axis)
ax.set_yticklabels(y_axis)
plt.title("TN3")
plt.xlabel("Horizontal")
plt.ylabel("Vertical")
plt.show()
```
